```python
# ➕ Add Total and Percentage
df['Total'] = df[['Mathematics', 'Science', 'English']].sum(axis=1)
df['Percentage'] = df['Total'] / 3

# ➕ Add Grade using conditions
def get_grade(pct):
    if pct >= 90:
        return 'A'
    elif pct >= 80:
        return 'B'
    elif pct >= 70:
        return 'C'
    elif pct >= 60:
        return 'D'
    else:
        return 'F'

df['Grade'] = df['Percentage'].apply(get_grade)
```
