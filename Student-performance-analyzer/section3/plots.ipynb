**Line Plot**

```python
path = "/content/drive/MyDrive/Colab Notebooks/students_cleaned.csv"
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the cleaned data
df = pd.read_csv(path)

# Step 2: Calculate average marks per subject
subjects = ['Mathematics', 'Science', 'English']
average_marks = [df[subject].mean() for subject in subjects]

# Step 3: Create line plot
plt.figure(figsize=(8, 5))
plt.plot(subjects, average_marks, marker='o', linestyle='-.', color='blue')

# Step 4: Customize the plot
plt.title('Average Marks per Subject')
plt.xlabel('Subject')
plt.ylabel('Average Marks')
plt.ylim(0, 100)  # Assuming max mark per subject is 100
plt.grid(True)

# Step 5: Show the plot
plt.tight_layout()
plt.show()
```

**Bar Chart**

```python
# Step 1: Count the number of students per grade
grade_counts = df['Grade'].value_counts().sort_index()  # Sort A to F

# Step 2: Create bar chart
plt.figure(figsize=(7, 5))
plt.bar(grade_counts.index, grade_counts.values, color=('orange','blue','red','green','yellow'),
        edgecolor='black')

# Step 3: Customize the plot
plt.title('Number of Students per Grade')
plt.xlabel('Grade')
plt.ylabel('Number of Students')
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Step 4: Add labels on top of bars
for i, value in enumerate(grade_counts.values):
    plt.text(i, value + 0.5, str(int(value)), ha='center', va='bottom')

# Step 5: Show plot
plt.tight_layout()
plt.show()
```

**Scatter Plot**

```python
# Define color for each grade
grade_colors = {
    'A': 'orange',
    'B': 'blue',
    'C': 'red',
    'D': 'green',
    'F': 'yellow'
}

# Map colors to each row based on the grade
colors = df['Grade'].map(grade_colors)

# Create scatter plot
plt.figure(figsize=(8, 5))
plt.scatter(df['Total'], df['Percentage'], c=colors, edgecolors='black')

# Customize plot
plt.title('Percentage vs. Total Marks (Color-coded by Grade)')
plt.xlabel('Total Marks')
plt.ylabel('Percentage (%)')
plt.grid(True)

# Add legend manually
import matplotlib.patches as mpatches
legend_handles = [mpatches.Patch(color=color, label=grade) for grade, color in grade_colors.items()]
plt.legend(handles=legend_handles, title='Grade')

# Show plot
plt.tight_layout()
plt.show()
```

```python
# Count number of students per grade
grade_counts = df['Grade'].value_counts().sort_index()

# Define colors for each grade
grade_colors = {
    'A': 'orange',
    'B': 'blue',
    'C': 'red',
    'D': 'green',
    'F': 'yellow'
}
colors = [grade_colors.get(grade, 'gray') for grade in grade_counts.index]

# Create Pie Chart
plt.figure(figsize=(7, 7))
plt.pie(
    grade_counts.values,
    labels=grade_counts.index,
    colors=colors,
    autopct='%1.1f%%',
    startangle=150,
    counterclock=False
)

plt.title("🎯 Grade Distribution of Students")
plt.legend(handles=legend_handles, title='Grade')
plt.axis('equal')  # Keep circle shape
plt.tight_layout()
plt.show()
```
